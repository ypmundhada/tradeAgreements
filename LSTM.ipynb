```python
import pandas as pd
import numpy as np
df=pd.read_csv("final.csv")
df.drop('Unnamed: 0',axis=1,inplace=True)
df
```

```python
y=df['gvc']
```

```python
X=df[['depth','AD','CP','CVD','ENV','ET','IPR','INV','LM','MIG','MOC','PP','ROR','SPS','STE','SER','SUB','TBT','TF','PTA','contig','ldistw','lgdp_o','lgdp_d']]
```

```python
df1=df.drop(['iso3_o','iso3_d','year'],axis=1)
df1
```

```python
columns=df1.columns[1:]
columns
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=2)
```

```python
X_train
```

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.layers import Embedding
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM
from tensorflow.keras.layers import Dense
from tensorflow.keras.layers import Bidirectional
from tensorflow.keras.layers import Dropout
from sklearn.preprocessing import RobustScaler, normalize
from sklearn.model_selection import GroupKFold, KFold
from sklearn.metrics import mean_absolute_error as mae
from tensorflow.keras.callbacks import EarlyStopping, ModelCheckpoint
from tensorflow.keras.callbacks import LearningRateScheduler, ReduceLROnPlateau
```

```python
tf.config.run_functions_eagerly(True)
```

```python
RS = RobustScaler()
X_train = RS.fit_transform(X_train)
X_test = RS.transform(X_test)
train = X_train.reshape(-1, 1, X_train.shape[-1])
test = X_test.reshape(-1, 1, X_train.shape[-1])
```

```python
model=Sequential()
model.add(Bidirectional(LSTM(1024,input_shape=(1,X_train.shape[-1]),return_sequences=True)))
model.add(Bidirectional(LSTM(512,return_sequences=True,activation='relu')))
model.add(Bidirectional(LSTM(256,return_sequences=True,activation='relu')))
model.add(Bidirectional(LSTM(128,return_sequences=True,activation='relu')))
model.add(Dense(1))
model.compile(optimizer='rmsprop',loss='mse')
```

```python
lrate = ReduceLROnPlateau(monitor="val_loss", factor=0.5, patience=10, verbose=1)
early = EarlyStopping(monitor="val_loss", patience=10, verbose=1, mode="min", restore_best_weights=True)
```

```python
model.fit(train, y_train, validation_data=(test, y_test), epochs=50, batch_size=128, callbacks=[lrate,early])
```

```python
results = []
predict = model.predict(test, verbose=0).squeeze() 
baseline_mse = np.mean(predict-y_test)
results.append({'feature':'BASELINE','mse':baseline_mse})
for k in range(X_train.shape[-1]):
    copy_column = test[:,:,k].copy()
    np.random.shuffle(test[:,:,k])
    predict = model.predict(test, verbose=0).squeeze() 
    mse = np.mean(predict-y_test)
    results.append({'feature':columns[k],'mse':mse})
    test[:,:,k] = copy_column
```

```python
import matplotlib.pyplot as plt
res = pd.DataFrame(results)
res = res.sort_values('mse')
plt.figure(figsize=(10,20))
plt.barh(np.arange(X_train.shape[-1]+1),res.mse)
plt.yticks(np.arange(X_train.shape[-1]+1),res.feature.values)
plt.ylim((-1,X_train.shape[-1]+1))
plt.plot([baseline_mse,baseline_mse],[-1,X_train.shape[-1]+1], '--', color='orange')
plt.ylabel('Feature',size=14)
plt.legend()
plt.show()
```

The most important provisions that impact gvcs as given by LSTM are:TBT, CP, LM, SPS, MIG, TF

The features given by PPML Lasso are: TBT, CP, SPS, MIG, TF, CVD, MIG, ROR, STE, SUB. We can see that PPML Lasso technique is a bit lenient in dealing with the features whereas LSTM penalizes the variables more.
