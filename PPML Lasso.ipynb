# PPML Lasso Regression

```python
import numpy as np
import gme as gme
df1=df1[df1['depth']>0]
df1['ldistw']=np.log(df1['distw'])
df1['lgdp_o']=np.log(df1['gdp_o'])
df1['lgdp_d']=np.log(df1['gdp_d'])
df1.drop(['distw','gdp_o','gdp_d'],axis=1,inplace=True)
df1
```

```python
df1.iso3_d.unique().size
```

```python
df1.dropna(inplace=True)
```

```python
data=gme.EstimationData(data_frame=df1,
                       imp_var_name='iso3_o',
                       exp_var_name='iso3_d',
                       trade_var_name='gvc',
                       year_var_name='year')
data
```

```python
model=gme.EstimationModel(estimation_data = data,
                         lhs_var = 'gvc',
                         rhs_var = ['ldistw','PTA','lgdp_o','lgdp_d','contig','AD','CP','CVD','ENV','ET','IPR','INV','LM','MIG','MOC','PP','ROR','SPS','STE','SER','SUB','TBT','TF'],
                         fixed_effects=[['iso3_d'],['iso3_o']])
estimate=model.estimate()
```

```python
model.results_dict["all"].summary()
```

```python
res=estimate['all']
coeff=res.params
coeff
```

```python
res.pvalues[:25]
```

Out of the 18 provisions:AD,CP,CVD,ENV,ET,IPR,INV,LM,MIG,MOC,PP,ROR,SPS,STE,SER,SUB,TBT,TF the ppml lasso model found that only 9 provisions had a significant impact on the GVCs (Global Value Chain). These significant variables are: CP,CVD,MIG,ROR,SPS,STE,SUB,TBT,TF with p values which are significant at 99% confidence level. These are the provisions which significantly impact GVC levels between two countries.
